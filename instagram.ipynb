```python
import pandas as pd 
import numpy as np
import plotly.express as px
```

```python
data = '/kaggle/input/top-instagram-influencers-data-cleaned/top_insta_influencers_data.csv'
```

```python
def load_dataset(path):
   return pd.read_csv(path)
df = load_dataset(data)
```

```python
def summarize_dataset(dataset):
    return dataset.info()
summarize_dataset(df)
```

```python
def new_post_avg_like(i):
    if str(i)[-1] == 'k':
        ready = str(i)[:-1]
        x = float(ready) * 1_000
        return x
    elif str(i)[-1] == 'm':
        ready = str(i)[:-1]
        y = float(ready) * 1_000_000
        return y
    else:
        return i
new_post_avg_like(df['new_post_avg_like'])
```

```python
def total_likes(i):
    if str(i)[-1] == 'k':
        ready = str(i)[:-1]
        x = float(ready) * 1_000
        return x
    elif str(i)[-1] == 'm':
        ready = str(i)[:-1]
        y = float(ready) * 1_000_000
        return y
    elif str(i)[-1] == 'b':
        ready = str(i)[:-1]
        z = float(ready) * 1_000_000_000
        return z
    else:
        return i
```

```python
def clean_dataset(dataset):
    dataset['posts'] = dataset['posts'].str.replace('k', '')
    dataset['posts'] = dataset['posts'].astype(float)*1000
    dataset['posts'] = dataset['posts'].astype(int)
    dataset['followers'] = dataset['followers'].str.replace('m', '')
    dataset['followers'] = dataset['followers'].astype(float)*1_000_000
    dataset['followers'] = dataset['followers'].astype(int)
    dataset['avg_likes'] = dataset['avg_likes'].apply(new_post_avg_like)
    dataset['60_day_eng_rate'] = dataset['60_day_eng_rate'].str.replace('%', '').astype(float)
    dataset['new_post_avg_like'] = dataset['new_post_avg_like'].apply(new_post_avg_like).astype(int)
    dataset['total_likes'] = dataset['total_likes'].map(total_likes).astype(int)
    dataset['country'] = dataset['country'].map(lambda x: "Other country" if type(x) == float else x)
    return dataset
clean_df = clean_dataset(df.copy(deep=True))
```

```python
top_rank = clean_df.sort_values(by='rank', axis=0, ascending=False).head(15)
fig = px.bar(top_rank, x='channel_info', y='rank', color='channel_info')
fig.show()
```
